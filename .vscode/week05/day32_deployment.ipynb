## 2. Deployment Process & Key Concepts

```python
print("""
HOW STREAMLIT CLOUD WORKS
==========================

1. READS requirements.txt
   Installs all packages listed
   Exact versions = reproducible builds
   No version = latest (risky — may break!)

2. RUNS app.py
   Same as: streamlit run app.py
   But on Streamlit's servers, not your laptop
   Auto-restarts if it crashes

3. SERVES via HTTPS
   Public URL — accessible anywhere
   No firewall, no port forwarding
   Free tier: unlimited public apps!

4. AUTO-DEPLOYS on git push
   Change app.py → git push → app updates!
   No manual redeployment needed
   Usually takes 1-2 minutes

REQUIREMENTS.TXT BEST PRACTICES
=================================
  ✅ Pin exact versions (==)     pandas==2.3.3
  ✅ Test locally first          streamlit run app.py
  ✅ Only include what you need  don't add unused packages
  ❌ Don't use >= alone          pandas>=1.0 (too vague)
  ❌ Don't include dev tools     jupyter, pytest etc.
  ❌ Don't add system packages   use packages.txt for those

GIT WORKFLOW FOR SEPARATE REPO
================================
  # First time setup (done today)
  git init
  git add .
  git commit -m "initial commit"
  git remote add origin <url>
  git push -u origin main

  # Future updates
  git add .
  git commit -m "update app"
  git push origin main
  # Streamlit Cloud auto-detects push and redeploys!

STREAMLIT CLOUD FREE TIER LIMITS
==================================
  - Unlimited public apps
  - 1GB RAM per app
  - Apps sleep after inactivity (wake on visit)
  - Private repos need paid plan
""")

# verify the deployed app would work locally too
import pandas as pd
import numpy as np
import pickle

print("Verifying model loads correctly...")
with open(r"C:\DS-AI-75d\.vscode\week05\project1\model\rf_model.pkl", "rb") as f:
    model = pickle.load(f)
with open(r"C:\DS-AI-75d\.vscode\week05\project1\model\features.pkl", "rb") as f:
    features = pickle.load(f)

# test prediction — same as what the deployed app would do
test = pd.DataFrame(
    [
        {
            "Pclass": 1,
            "Age": 25,
            "FareLog": np.log1p(100),
            "FamilySize": 1,
            "IsAlone": 1,
            "HasCabin": 1,
            "Sex_encoded": 1,
            "Title_Mr": 0,
            "Title_Mrs": 1,
            "Title_Miss": 0,
            "Title_Master": 0,
        }
    ]
)

prob = model.predict_proba(test)[0][1]
print(f"\nTest prediction (1st class Mrs, 25):")
print(f"  Survival probability: {prob*100:.1f}%")
print(f"  Prediction: {'Survived ✅' if prob >= 0.5 else 'Died ❌'}")
print(f"\nModel features: {features}")
print(f"\nDeployment verified! ✅")
```
