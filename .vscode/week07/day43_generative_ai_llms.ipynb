## 2. What is Generative AI & LLMs?

```python
print("=" * 55)
print("       WHAT IS GENERATIVE AI & LLMs?")
print("=" * 55)
print("""
GENERATIVE AI:
  AI that creates NEW content — text, images, code, audio
  Learns patterns from data → generates similar new data
  Examples: ChatGPT, Claude, Gemini, DALL-E, Midjourney

LLM = Large Language Model:
  A neural network trained on massive text data
  Learns to predict the next token given previous tokens
  'Large' = billions of parameters (GPT-3: 175B, GPT-4: ~1T)

HOW LLMs GENERATE TEXT:
  1. Take a prompt: "The cat sat on the"
  2. Predict probability of next token:
       "mat"  → 0.45
       "floor" → 0.20
       "roof"  → 0.15
       ...
  3. Sample from distribution → pick "mat"
  4. Append "mat" → repeat from step 2
  5. Continue until [END] token or max length

GPT vs BERT:
  BERT  → Encoder  → READS  → understands text
  GPT   → Decoder  → WRITES → generates text

  GPT = Generative Pretrained Transformer
  Trained to predict next token — left to right only
  No peeking at future tokens (causal/masked attention)

KEY LLMs TODAY:
  GPT-4      → OpenAI   → most capable, closed source
  Claude     → Anthropic → safe, helpful, honest
  Gemini     → Google   → multimodal
  LLaMA 3    → Meta     → open source, powerful
  Mistral 7B → Mistral  → efficient open source
""")
```

## 3. Prompt Engineering

```python
print("=" * 55)
print("       PROMPT ENGINEERING")
print("=" * 55)
print("""
PROMPT ENGINEERING = designing inputs to get better outputs

THREE KEY TECHNIQUES:

1. ZERO-SHOT PROMPTING
   No examples — just ask directly
   "Classify this review as positive or negative:
    'The food was amazing!' → "

2. FEW-SHOT PROMPTING
   Give examples before asking
   "Classify sentiment:
    'Great film!' → POSITIVE
    'Terrible movie' → NEGATIVE
    'Loved every minute' → "

3. CHAIN-OF-THOUGHT (CoT)
   Ask the model to reason step by step
   "Q: Roger has 5 balls. He buys 2 more cans
       of 3 balls each. How many does he have?
    A: Roger starts with 5. 2 cans × 3 = 6 more.
       5 + 6 = 11. The answer is 11."

PROMPT ENGINEERING TIPS:
  ✅ Be specific and clear — vague prompts = vague outputs
  ✅ Give examples (few-shot) for complex tasks
  ✅ Ask for step-by-step reasoning for maths/logic
  ✅ Specify format: 'Answer in JSON', 'In 3 bullet points'
  ✅ Give role: 'You are an expert data scientist...'
  ✅ Iterate — prompt engineering is experimental!

SYSTEM PROMPTS:
  Instructions given to the model before the conversation
  Sets behaviour, tone, constraints
  "You are a helpful assistant that answers only
   in Python code."
""")
```

## 4. Text Generation with GPT-2

```python
print("=" * 55)
print("       TEXT GENERATION WITH GPT-2")
print("=" * 55)

# load GPT-2
print("Loading GPT-2...")
generator = pipeline('text-generation', model='gpt2', device=-1)
print("Loaded! ✅\n")

prompts = [
    "Artificial intelligence will",
    "The best way to learn data science is",
    "In the future, robots will",
]

for prompt in prompts:
    print(f"PROMPT: '{prompt}'")
    result = generator(
        prompt,
        max_new_tokens=40,
        num_return_sequences=1,
        temperature=0.8,
        do_sample=True,
        pad_token_id=50256
    )
    generated = result[0]['generated_text']
    print(f"OUTPUT: {generated}")
    print("-" * 55)
```

## 5. Generation Parameters — Temperature & Sampling

```python
print("=" * 55)
print("       GENERATION PARAMETERS")
print("=" * 55)
print("""
KEY PARAMETERS that control text generation:

TEMPERATURE:
  Controls randomness of output
  Low  (0.1) → very focused, repetitive, predictable
  High (1.5) → creative, random, sometimes incoherent
  Default    → 1.0 (balanced)

TOP-K SAMPLING:
  At each step, only consider top K most likely tokens
  top_k=50 → sample from top 50 tokens only
  Prevents very unlikely tokens being chosen

TOP-P (NUCLEUS) SAMPLING:
  Consider smallest set of tokens whose probability sums to P
  top_p=0.9 → tokens covering 90% of probability mass
  More dynamic than top_k — adapts to each distribution

MAX NEW TOKENS:
  Maximum number of tokens to generate
  Does not include the prompt length

GREEDY vs SAMPLING:
  do_sample=False → always pick highest probability token
                    (deterministic, often repetitive)
  do_sample=True  → sample from distribution
                    (creative, varied)
""")

prompt = "Data science is"
tokenizer_gpt2 = GPT2Tokenizer.from_pretrained('gpt2')
model_gpt2     = GPT2LMHeadModel.from_pretrained('gpt2')
model_gpt2.eval()

configs = [
    ("Greedy (temp=1.0, no sample)", dict(do_sample=False, max_new_tokens=30)),
    ("Creative (temp=1.2, sample)",  dict(do_sample=True, temperature=1.2, max_new_tokens=30, top_p=0.9)),
    ("Focused (temp=0.3, sample)",   dict(do_sample=True, temperature=0.3, max_new_tokens=30, top_k=50)),
]

inputs = tokenizer_gpt2(prompt, return_tensors='pt')
print(f"PROMPT: '{prompt}'\n")
for name, params in configs:
    with torch.no_grad():
        output = model_gpt2.generate(
            inputs['input_ids'],
            pad_token_id=50256,
            **params
        )
    text = tokenizer_gpt2.decode(output[0], skip_special_tokens=True)
    print(f"[{name}]")
    print(f"  {text}")
    print()
```
