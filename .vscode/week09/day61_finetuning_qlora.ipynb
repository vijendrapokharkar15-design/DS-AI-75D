## What Was Done Today (Google Colab — T4 GPU)

**Platform:** Google Colab (free tier, Tesla T4 GPU, 15GB VRAM)

**Base Model:** unsloth/Phi-3-mini-4k-instruct (3.8B parameters)

**Fine-Tuning Method:** QLoRA (4-bit quantized base) + LoRA adapters via Unsloth

**Dataset:** 10 custom instruction-following examples covering DS/AI topics
(overfitting, precision/recall, transformers, RAG, LoRA, attention mechanism, etc.)

- Rank (r): 16
- Target modules: q_proj, k_proj, v_proj, o_proj, gate_proj, up_proj, down_proj

**Training Config:**
- Epochs: 3
- Batch size: 2 (gradient accumulation: 4, effective batch: 8)
- Learning rate: 2e-4
- Optimizer: adamw_8bit
- Training steps: 6

**Result:** Model correctly answered RAG, LoRA, and other DS/AI questions
after fine-tuning, demonstrating successful knowledge adaptation.
