## 2. Indexing & Filtering
Three ways to select data in Pandas — master all three.

### 2.1 Column Selection

```python
# Single column → returns a Series
age = df["Age"]
print("Type:", type(age))
print("Shape:", age.shape)

# Multiple columns → returns a DataFrame
subset = df[["Name", "Age", "Sex", "Survived"]]
print("\nType:", type(subset))
print("Shape:", subset.shape)
subset.head()
```

### 2.2 loc — Label Based Selection

```python
# loc uses ROW LABELS and COLUMN NAMES
# Format: df.loc[row, column]

# Single row
print("=== Single row ===")
print(df.loc[0])

# Specific value
print("\n=== Single value ===")
print(df.loc[0, "Age"])

# Range of rows + specific columns
print("\n=== Rows 0-4, selected columns ===")
df.loc[0:4, ["Name", "Age", "Sex", "Survived"]]
```

### 2.3 iloc — Position Based Selection

```python
# iloc uses ROW NUMBERS and COLUMN NUMBERS
# Format: df.iloc[row_position, col_position]

# First row by position
print("=== First row ===")
print(df.iloc[0])

# Specific position
print("\n=== Row 0, Column 1 (Survived) ===")
print(df.iloc[0, 1])

# Range — remember EXCLUSIVE like NumPy
print("\n=== First 5 rows, first 4 columns ===")
df.iloc[0:5, 0:4]
```

### Key Difference: loc vs iloc
| | loc | iloc |
|---|---|---|
| Uses | Labels/Names | Numbers/Positions |
| End of range | INCLUSIVE | EXCLUSIVE |
| Example | df.loc[0:4] = 5 rows | df.iloc[0:4] = 4 rows |

### 2.4 Boolean Filtering — The Most Used Skill in Real DS Work

```python
# Single condition
survivors = df[df["Survived"] == 1]
first_class = df[df["Pclass"] == 1]
adults = df[df["Age"] >= 18]

print("=== Survival counts ===")
print(f"Total passengers:  {len(df)}")
print(f"Survivors:         {len(survivors)}")
print(f"Non-survivors:     {len(df) - len(survivors)}")
print(f"Survival rate:     {len(survivors)/len(df)*100:.1f}%")

print("\n=== Multiple conditions ===")
# AND — both conditions must be true
first_class_survivors = df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
print(f"First class survivors: {len(first_class_survivors)}")

# OR — either condition can be true
children_or_elderly = df[(df["Age"] < 12) | (df["Age"] > 60)]
print(f"Children or elderly:   {len(children_or_elderly)}")

# NOT
third_class = df[~(df["Pclass"] == 3)]
print(f"Not third class:       {len(third_class)}")
```

## 3. GroupBy Analysis — Who Survived and Why?
Split → Apply → Combine pattern.

### Question 1: What was the overall survival rate?

```python
overall = df["Survived"].mean()
total = len(df)
survived = df["Survived"].sum()

print(f"Total passengers:  {total}")
print(f"Total survivors:   {int(survived)}")
print(f"Overall survival rate: {overall:.1%}")
```

### Question 2: Did passenger class determine survival?

```python
class_survival = df.groupby("Pclass")["Survived"].mean()
class_count = df.groupby("Pclass")["Survived"].count()

result = pd.DataFrame(
    {
        "Survival Rate": class_survival.map("{:.1%}".format),
        "Total Passengers": class_count,
    }
)
print(result)
```

### Question 3: Did gender determine survival?

```python
gender_survival = df.groupby("Sex")["Survived"].mean()
gender_count = df.groupby("Sex")["Survived"].count()

result = pd.DataFrame(
    {
        "Survival Rate": gender_survival.map("{:.1%}".format),
        "Total Passengers": gender_count,
    }
)
print(result)
```

### Question 4: What was the age profile of survivors vs non-survivors?

```python
age_profile = df.groupby("Survived")["Age"].agg(["mean", "median", "std"])
age_profile.index = ["Did not survive", "Survived"]
print(age_profile.round(2))
```

### Question 5: Which class + gender combination had best/worst survival?

```python
pivot = df.groupby(["Pclass", "Sex"])["Survived"].mean().unstack()
print("Survival rate by class and gender:")
print(pivot.map("{:.1%}".format))
```

## 4. DataFrame Merging

```python
# Create class statistics to merge back
class_stats = (
    df.groupby("Pclass")
    .agg(
        avg_survival=("Survived", "mean"),
        avg_fare=("Fare", "mean"),
        passenger_count=("PassengerId", "count"),
    )
    .reset_index()
)

print("Class statistics table:")
print(class_stats.round(2))
```

```python
# Merge class stats back into main DataFrame
df_enriched = pd.merge(df, class_stats, on="Pclass", how="left")

print("Shape before merge:", df.shape)
print("Shape after merge: ", df_enriched.shape)
print("\nSample — each passenger now has their class statistics:")
df_enriched[["Name", "Pclass", "Survived", "avg_survival", "avg_fare"]].head(8)
```

### The Four Join Types

```python
# Create two small tables to demonstrate all join types
passengers = pd.DataFrame(
    {
        "PassengerId": [1, 2, 3, 4, 5],
        "Name": ["Allen", "Braund", "Cumings", "Heikkinen", "Futrelle"],
        "Pclass": [1, 3, 1, 3, 1],
    }
)

ticket_prices = pd.DataFrame({"Pclass": [1, 2, 3], "AvgFare": [84.15, 20.66, 13.68]})

# INNER — only matching rows from both
inner = pd.merge(passengers, ticket_prices, on="Pclass", how="inner")
print("=== INNER JOIN ===")
print(inner)

# LEFT — all from left, matching from right
left = pd.merge(passengers, ticket_prices, on="Pclass", how="left")
print("\n=== LEFT JOIN ===")
print(left)
```

```python
# RIGHT and OUTER joins — shows NaN when no match
right = pd.merge(passengers, ticket_prices, on="Pclass", how="right")
print("=== RIGHT JOIN ===")
print(right)
# Notice: Pclass 2 appears with NaN for passenger details
# because no passenger in our small table has Pclass 2

outer = pd.merge(passengers, ticket_prices, on="Pclass", how="outer")
print("\n=== OUTER JOIN ===")
print(outer)
```
